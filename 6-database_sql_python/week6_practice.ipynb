## Problem 1
### List the case number, type of crime and community area for all crimes in community area number 18.

```python
%%sql
select C.case_number, C.primary_type, N.community_area_name
from CHICAGO_CRIME_DATA as C
inner join CENSUS_DATA as N on C.community_area_number = N.community_area_number
where C.community_area_number = 18
```

## Problem 2
### List all crimes that took place at a school. Include case number, crime type and community name.

```python
%%sql
select C.case_number, C.primary_type, N.community_area_name
from CHICAGO_CRIME_DATA as C
right join CENSUS_DATA as N on C.community_area_number = N.community_area_number
where C.location_description like '%SCHOOL%'
```

## Problem 3
### For the communities of Oakland, Armour Square, Edgewater and CHICAGO list the associated community_area_numbers and the case_numbers.

```python
select C.case_number, N.community_area_number
from CHICAGO_CRIME_DATA as C
full outer join CENSUS_DATA as N on C.community_area_number = N.community_area_number
where N.community_area_name in ('Oakland','Armour Square','Edgewater','CHICAGO')
```
